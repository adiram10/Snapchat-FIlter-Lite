```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_eye.xml")
```

```python
img=plt.imread("/content/Adi6.jpg")
plt.imshow(img)
```

```python
img.shape
```

```python
img1=img.copy()
```

```python
eye=eye_cascade.detectMultiScale(img)[0]
print(eye)
```

```python
eye1=eye_cascade.detectMultiScale(img)[1]
print(eye1)
```

```python
eye_x,eye_y,eye_w, eye_h= eye
```

```python
eye_x1,eye_y1,eye_w1, eye_h1= eye1
```

```python
img = cv2.rectangle(img, (eye_x, eye_y), (eye_x + eye_w, eye_y + eye_h), (255,255,255), 5 ) 
```

```python
img = cv2.rectangle(img, (eye_x1, eye_y1), (eye_x1 + eye_w1, eye_y1 + eye_h1), (255,255,255), 5 ) 
```

```python
plt.imshow(img)
```
